```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
def xy_split(df):
    
    return df.drop(columns= 'prognosis'), df.prognosis
```

```python
test = pd.read_csv('disease.csv')
```

```python
test.head()
```

```python
X_test, y_test = xy_split(test)
```

```python
test.shape
```

```python
df = pd.read_csv('Disease2.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.prognosis.value_counts()
```

```python
train, val = train_test_split(df, train_size = 0.9,
                                       random_state = 42,
                                       stratify = df['prognosis'])
```

```python
train.shape, val.shape
```

```python
X_train, y_train = xy_split(train)
X_val, y_val = xy_split(val)
```

```python
X_train.head()
```

```python
X_train.shape, X_val.shape
```

# Logistics

```python
lm = LogisticRegression(class_weight = 'balanced')

lm.fit(X_train, y_train)
```

## Train Results

```python
y_train_res = pd.DataFrame({'actual': y_train,
                            'preds': lm.predict(X_train)})

y_train_res.head()
```

```python
print(classification_report(y_train_res.actual, y_train_res.preds))
```

## Val Results

```python
y_val_res = pd.DataFrame({'actual': y_val,
                            'preds': lm.predict(X_val)})

y_val_res.head()
```

```python
print(classification_report(y_val_res.actual, y_val_res.preds))
```

## Test Results

```python
y_test_res = pd.DataFrame({'actual': y_test,
                            'preds': lm.predict(X_test)})

y_test_res.head()
```

```python
print(classification_report(y_test_res.actual, y_test_res.preds))
```

# Random Forest

```python
rf = RandomForestClassifier(class_weight = 'balanced')
rf.fit(X_train, y_train)
```

## Train Results

```python
y_train_res = pd.DataFrame({'actual': y_train,
                            'preds': rf.predict(X_train)})

y_train_res.head()
```

```python
print(classification_report(y_train_res.actual, y_train_res.preds))
```

## Val Results

```python
y_val_res = pd.DataFrame({'actual': y_val,
                            'preds': rf.predict(X_val)})

y_val_res.head()
```

```python
print(classification_report(y_val_res.actual, y_val_res.preds))
```
